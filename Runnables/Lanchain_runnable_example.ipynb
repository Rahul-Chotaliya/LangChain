```python
import random

class NakliLLm:
    def __init__(self):
        print("LLm Created!!!!")
    
    def predict(self,prompt):
        response_list = [
            "Virat Kohli is the Greatest of all time",
            "Hardik Pandya is the 1st ranker of All time all rounder in ICC",
            "Rishabh Pant is trying to grow-up with his skills"
        ]
        
        return {"response": random.choice(response_list)}
```

```python
llm = NakliLLm()
print(llm)
llm.predict("Rahul Chotaliya")
```

```python
class NakliPromptTemplate:
    def __init__(self,template,input_variable):
        self.template = template
        self.input_variable = input_variable
    
    def format(self,input_dict):
        return self.template.format(**input_dict)
```

```python
template = NakliPromptTemplate(
    template="Write a {length} poem about {topic}",
    input_variable=["length","topic"]
)
```

```python
prompt = template.format({"length":"short","topic":"Virat Kohli"})
```

```python
llm = NakliLLm()
```

```python
llm.predict(prompt)
```

```python
class NakliLLMChain:
    def __init__(self,llm,prompt):
        self.llm = llm
        self.prompt = prompt
    
    def run(self,input_dict):
        final_prompt = self.prompt.format(input_dict)
        result = self.llm.predict(final_prompt)
        return result["response"]
```

```python
chain = NakliLLMChain(llm,template)
```

```python
chain.run({"length":"short","topic":"Hardik Pandya"})
```
