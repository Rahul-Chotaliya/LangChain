```python
from langchain_google_genai import GoogleGenerativeAIEmbeddings
from langchain_community.vectorstores import Chroma
```

```python
from langchain_core.documents import Document

doc1 = Document(
    page_content="Virat Kohli is one of the greatest cricketers of all time. Known for his aggressive batting style and consistency, he has set numerous records in international cricket.", 
    metadata={"Team": "Royal Challengers Bangalore", "source": "doc1.txt"}
)
doc2 = Document(
    page_content="Rohit Sharma is a prolific Indian cricketer, renowned for his ability to score big centuries. He holds the record for the highest individual score in One Day Internationals (ODIs) with 264 runs.", 
    metadata={"Team": "Mumbai Indians", "source": "doc2.txt"}
)   

doc3 = Document(
    page_content = "Hardik Pandya is an Indian cricketer known for his all-round abilities. He has been a key player for the Indian cricket team, contributing with both bat and ball in various formats of the game.",
    metadata={"Team": "Chennai Super Kings", "source": "doc3.txt"}
)

doc4 = Document(
    page_content = "Jasprit Bumrah is an Indian fast bowler known for his unique bowling action and ability to bowl yorkers consistently. He has been a crucial part of the Indian cricket team, especially in limited-overs formats.",
    metadata={"Team": "Mumbai Indians", "source": "doc4.txt"}
)

doc5 = Document(
    page_content = "MS Dhoni is a legendary Indian cricketer and former captain of the Indian national team. He is widely regarded as one of the greatest wicket-keeper batsmen in cricket history, known for his calm demeanor and finishing abilities in high-pressure situations.",
    metadata={"Team": "Chennai Super Kings", "source": "doc5.txt"}
)

docs = [doc1, doc2, doc3, doc4, doc5]






```

```python
from langchain_google_genai import GoogleGenerativeAIEmbeddings

embeddings = GoogleGenerativeAIEmbeddings(
    model="gemini-embedding-001"
)
```

```python
vector_store = Chroma.from_documents(
    documents=docs,
    embedding=embeddings,
    collection_name="cricket_players_db",
    persist_directory="./chroma_db"
)
```

```python
vector_store.add_documents(docs)
```

```python
vector_store.get(include=["embeddings", "metadatas", "documents"])
```

```python
vector_store.similarity_search(
    query = "Which cricketer is known for his aggressive batting style?",
    k = 2   
)
```

```python
vector_store.similarity_search(
    query = "Who among these are a bowler?",
    k = 2
)
```

```python
vector_store.similarity_search_with_score(
    query = "Who among these are a bowler?",
    k = 1,
)
```

```python
vector_store.similarity_search_with_score(
    query = "highest score in IPL?"
)
```
