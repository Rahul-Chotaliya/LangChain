```python
from langchain_core.documents import Document

documents = [
    Document(
        page_content="Multi Query Retrieval is a technique used to improve search results by generating multiple variations of a user query instead of relying on a single query.",
        metadata={"source": "intro"}
    ),
    Document(
        page_content="The main idea is that a single query may not capture all possible ways information is expressed. By creating multiple queries, the system can retrieve a broader and more relevant set of documents.",
        metadata={"source": "intuition"}
    ),
    Document(
        page_content="In Multi Query Retrieval, a language model is typically used to generate different reformulations of the original query. These variations may include synonyms, rephrased questions, or related concepts.",
        metadata={"source": "query_generation"}
    ),
    Document(
        page_content="Each generated query is then used to perform a separate search in the vector database. The results from all queries are combined into a single pool of candidate documents.",
        metadata={"source": "retrieval_process"}
    ),
    Document(
        page_content="After retrieving documents from multiple queries, duplicate or highly similar documents are removed to ensure diversity and reduce redundancy.",
        metadata={"source": "deduplication"}
    ),
    Document(
        page_content="For example, for the query 'benefits of exercise', the system might generate queries like 'advantages of physical activity', 'why is exercise important', and 'health benefits of working out'.",
        metadata={"source": "example_queries"}
    ),
    Document(
        page_content="This approach helps retrieve documents that might be missed by a single query due to differences in wording, terminology, or phrasing.",
        metadata={"source": "benefit"}
    ),
    Document(
        page_content="Multi Query Retrieval is commonly used in Retrieval-Augmented Generation (RAG) systems to improve recall and ensure the language model has access to more comprehensive context.",
        metadata={"source": "applications"}
    ),
    Document(
        page_content="It can also be combined with techniques like Maximum Marginal Relevance (MMR) to further improve diversity and relevance in the final selected documents.",
        metadata={"source": "combination"}
    ),
]
```

```python
from langchain_google_genai import GoogleGenerativeAIEmbeddings
from langchain_community.vectorstores import FAISS
```

```python
embeddings = GoogleGenerativeAIEmbeddings(
    model="gemini-embedding-001"
)
vector_store = FAISS.from_documents(documents, embeddings)
```

```python
similarity_retriever = vector_store.as_retriever(search_kwargs={"k": 5}, search_type="similarity")
```

```python
from langchain.retrievers.multi_query import MultiQueryRetriever
from langchain_google_genai import GoogleGenerativeAI

multiquery_retriever = MultiQueryRetriever(
    retriever=vector_store.as_retriever(search_kwargs={"k": 5}),
    llm = GoogleGenerativeAI(model="gemini-1.5-flash-preview")
)
```

```python
query = "What are the benefits of exercise?"
```

```python
similarity_result = similarity_retriever.invoke(query)
multiquery_result = multiquery_retriever.invoke(query)
```

```python
for i , doc in enumerate(similarity_result):
    print(f"Similarity Result {i+1}: {doc.page_content} (source: {doc.metadata['source']})")

for i , doc in enumerate(multiquery_result):
    print(f"Multi-Query Result {i+1}: {doc.page_content} (source: {doc.metadata['source']})")
```
