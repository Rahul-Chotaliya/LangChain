```python
from langchain_community.vectorstores import Chroma
from langchain_google_genai import GoogleGenerativeAIEmbeddings
from langchain_core.documents import Document
```

```python
documents = [
    Document(page_content="The capital of France is Paris.", metadata={"source": "source1"}),
    Document(page_content="The capital of Germany is Berlin.", metadata={"source": "source2"}),
    Document(page_content="The capital of Italy is Rome.", metadata={"source": "source3"}),
    Document(page_content="The capital of Spain is Madrid.", metadata={"source": "source4"}),
    Document(page_content="The capital of Portugal is Lisbon.", metadata={"source": "source5"})
    
]
```

```python
embeddings = GoogleGenerativeAIEmbeddings(
    model="gemini-embedding-001"
)

vector_store = Chroma.from_documents(documents, embeddings,collection_name="capitals")
```

```python
retriever = vector_store.as_retriever(search_kwargs={"k": 3})
```

```python
query = "Rome"
results = retriever.invoke(query)
```

```python
for i,ddoc in enumerate(results):
    print(f"Result {i+1}:")
    print(f"Content: {ddoc.page_content}")
    print(f"Metadata: {ddoc.metadata}")
    print()
```
