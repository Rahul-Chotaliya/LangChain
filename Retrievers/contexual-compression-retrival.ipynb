```python
from langchain_core.documents import Document

documents = [
    Document(
        page_content="Contextual Compression Retrieval is a technique used to reduce the amount of retrieved text while keeping only the most relevant information for a query.",
        metadata={"source": "intro"}
    ),
    Document(
        page_content="The main goal is to remove unnecessary or irrelevant parts of documents so that the language model receives concise and useful context.",
        metadata={"source": "goal"}
    ),
    Document(
        page_content="In many retrieval systems, documents may contain extra information that is not directly related to the query. Contextual compression filters or shortens this content.",
        metadata={"source": "problem"}
    ),
    Document(
        page_content="Contextual Compression works by first retrieving documents using a standard retriever and then applying a compressor to refine the results.",
        metadata={"source": "process"}
    ),
    Document(
        page_content="A compressor can be a language model or a filtering algorithm that extracts only the relevant sentences or sections from each document.",
        metadata={"source": "compressor"}
    ),
    Document(
        page_content="For example, if a document contains a full article about machine learning but the query is about neural networks, the compressor will keep only the neural network-related parts.",
        metadata={"source": "example"}
    ),
    Document(
        page_content="This approach helps reduce token usage, improve response quality, and avoid overwhelming the language model with unnecessary information.",
        metadata={"source": "benefits"}
    ),
    Document(
        page_content="Contextual Compression is commonly used in Retrieval-Augmented Generation (RAG) systems to optimize the context passed to the model.",
        metadata={"source": "applications"}
    ),
    Document(
        page_content="It can be combined with techniques like Multi Query Retrieval and Maximum Marginal Relevance to first improve recall and diversity, and then compress the final results.",
        metadata={"source": "combination"}
    ),
]
```

```python
from langchain_google_genai import GoogleGenerativeAI,GoogleGenerativeAIEmbeddings
from langchain_community.vectorstores import FAISS
embeddings = GoogleGenerativeAIEmbeddings(
    model="gemini-embedding-001"
)
vectore_store = FAISS.from_documents(documents, embeddings)
```

```python
best_retriever = vectore_store.as_retriever(search_kwargs={"k": 5})
```

```python
# llm = GoogleGenerativeAI(model = "gemini-3-flash-preview")

from langchain.retrievers.document_compressors import LLMChainExtractor
from langchain.retrievers.contextual_compression import ContextualCompressionRetriever

llm = GoogleGenerativeAI(model = "gemini-3-flash-preview")
compressor = LLMChainExtractor.from_llm(llm)
```

```python
compression_retriever = ContextualCompressionRetriever(
    base_retriever=best_retriever,
    base_compressor=compressor
) 
```

```python
query = "What is photosynthesis?"

compressed_results = compression_retriever.invoke(query)
```

```python
for i, doc in enumerate(compressed_results):
    print(f"Result {i+1}:")
    print(f"Content: {doc.page_content}")
    print(f"Metadata: {doc.metadata}")
    print("-" * 50)
```
