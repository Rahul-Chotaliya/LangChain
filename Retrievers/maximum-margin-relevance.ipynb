```python
from langchain_core.documents import Document


documents = [
    Document(
        page_content="Maximum Marginal Relevance (MMR) is a method used to select results that are both relevant to a query and diverse among themselves.",
        metadata={"source": "intro"}
    ),
    Document(
        page_content="The key idea behind MMR is to avoid redundancy. Instead of returning many similar results, it balances relevance to the query with novelty compared to already selected items.",
        metadata={"source": "intuition"}
    ),
    Document(
        page_content="MMR works by iteratively selecting documents. At each step, it chooses the document that maximizes a score combining query relevance and dissimilarity to previously selected documents.",
        metadata={"source": "process"}
    ),
    Document(
        page_content="The MMR formula is: MMR = lambda * similarity(query, document) - (1 - lambda) * max similarity(document, selected_documents). Lambda controls the trade-off between relevance and diversity.",
        metadata={"source": "formula"}
    ),
    Document(
        page_content="If lambda is close to 1, MMR prioritizes relevance. If lambda is close to 0, it prioritizes diversity. A common choice is 0.5 to balance both.",
        metadata={"source": "lambda_effect"}
    ),
    Document(
        page_content="For example, when searching for 'machine learning', MMR ensures results include different subtopics like supervised learning, deep learning, and reinforcement learning instead of repeating similar articles.",
        metadata={"source": "example"}
    ),
    Document(
        page_content="MMR is widely used in information retrieval, search engines, and retrieval-augmented generation (RAG) systems to improve the quality and coverage of results.",
        metadata={"source": "applications"}
    ),
    Document(
        page_content="In vector databases, MMR is often applied after similarity search to re-rank the top results and remove redundant chunks while preserving useful variety.",
        metadata={"source": "vector_db_usage"}
    ),
]
```

```python
from langchain_community.vectorstores import FAISS
from langchain_google_genai import GoogleGenerativeAIEmbeddings

embeddings = GoogleGenerativeAIEmbeddings(
    model="gemini-embedding-001"
)
```

```python
vector_store = FAISS.from_documents(documents, embeddings)
```

```python
retriever = vector_store.as_retriever(
    search_kwargs={"k": 3, "lambda_mult": 0.1},
    search_type="mmr"
)
```

```python
query = "cons of mmr"
results = retriever.invoke(query)
```

```python
for i,doc in enumerate(results):
    print(f"Result {i+1}:")
    print(f"Content: {doc.page_content}")
    print(f"Metadata: {doc.metadata}")
    print()
```
