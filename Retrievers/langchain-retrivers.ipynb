```python
from langchain_community.retrievers import WikipediaRetriever
```

```python
retriever = WikipediaRetriever(top_k_results=2, leng="en")
```

```python
query = "the geopolitical history of india and pakistan from the perspective of chinese"
docs = retriever.invoke(query)
```

```python
docs
```

```python
for i , doc in enumerate(docs):
    print(f"Document {i+1}:")
    print(f"Title: {doc.metadata['title']}")
    print(f"Content: {doc.page_content[:500]}...")  # Print the first 500 characters of the content
    print("\n")
```
